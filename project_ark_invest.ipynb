### Set the variables for the Alpaca API and secret keys. Import, slice and clean datas and set it to the dataframe.

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
alpaca_api_key 
alpaca_secret_key 
# Create the Alpaca tradeapi.REST object
# YOUR CODE HERE
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
tickers = ["ARKK", "ARKQ", "ARKW", "ARKG", "ARKF", "SPY"]
# Set timeframe to 1D 
# YOUR CODE HERE
timeframe = "1D"

# Format current date as ISO format
# Set both the start and end date at the date of your prior weekday 
# This will give you the closing price of the previous trading day
# Alternatively you can use a start and end date of 2020-08-07
start_date = pd.Timestamp("2019-01-01", hour=None,minute=None, second=None, tz="America/Chicago").isoformat()
end_date = pd.Timestamp("2022-02-14", hour=None,minute=None, second=None, tz="America/Chicago").isoformat()
```

```python
ark_df = alpaca.get_barset(
    tickers,
    timeframe, 
    start=start_date,
    end=end_date,
    limit=1000
).df

# Review the first 5 rows of the Alpaca DataFrame
# YOUR CODE HERE
display(ark_df.head())
display(ark_df.tail())
```

```python
ark_df = ark_df.dropna()
ark_df.head()
#prices_ark_close_dfd = prices_ark_close_df.drop(prices_ark_close_df.iloc[:, 1:"close"], axis=1, inplace=True)
#prices_ark_close_dfd
```

```python
ark_df.columns = ['_'.join(x) for x in ark_df.columns]
```

```python
ark_df_close = ark_df.loc[:, ark_df.columns.str.contains("_close")].rename_axis("Date")
ark_df_close.head()
```

## Use the Pandas pct_change function together with dropna to create the daily returns DataFrame.

```python
ark_df_daily_return = ark_df_close.pct_change().dropna()
ark_df_daily_return
```

# Analyze the Risk-Return Profile
#### To determine the overall risk of an asset or portfolio, quantitative analysts and investment managers consider not only its risk metrics but also its risk-return profile. After all, if you have two portfolios that each offer a 10% return but one has less risk, you’d probably invest in the smaller-risk portfolio. For this reason, you need to consider the Sharpe ratios for each portfolio.

##### Step 1: Use the daily return DataFrame to calculate the annualized average return data for the five ETF's and for the S&P 500. Use 252 for the number of trading days. Review the annualized average returns, sorted from lowest to highest.

```python
ark_df_annual_return = ark_df_daily_return.mean()*252
ark_df_annual_return.sort_values()
```

##### Step 2: Calculate the Sharpe ratios for the five ETF's and for the SPY. To do that, divide the annualized average return by the annualized standard deviation for each. Review the resulting Sharpe ratios, sorted from lowest to highest.¶

```python
sharpe_ratios_ark_df = ark_df_annual_return / ark_df_annual_std_dev
sharpe_ratios_ark_df.sort_values()
```

##### Step 3: Visualize the Sharpe ratios for the five ETF's and for the SPY in a bar chart. 

```python
sharpe_ratios_ark_df.plot.bar(figsize=(10,5), title="Sharpe Ratios")
```

# Diversifying the Portfolio¶


### Calculate the variance of the SPY by using a 60-day rolling window and visualize the last five rows.


```python
spy_60_var = ark_df_daily_return.loc[:, "SPY_close"].rolling(window=60).var()                                                  
spy_60_var.tail()
```

##### ARKF - Step 1: Calculate the covariance of ARKF by using the 60-day rolling window, the daily return data, and the SPY_close returns, . Review the last five rows of the covariance of ARKF.¶


```python
arkf_60_cov = ark_df_daily_return["ARKF_close"].rolling(window=60).cov( ark_df_daily_return["SPY_close"])
arkf_60_cov.tail()
```

##### ARKF  - Step 2: Calculate the Beta of ARKF based on the 60-day rolling covariance. To do that, divide the covariance of ARKF by the variance of the SPY.

```python
arkf_60_beta = arkf_60_cov / spy_60_var
arkf_60_beta.fillna(0)
```

##### ARKF  - Step 3: Use the Pandas mean function to calculate the average value of the 60-day rolling beta of ARKF.¶

```python
arkf_60_beta_ave = arkf_60_beta.mean()
arkf_60_beta_ave
print(f"The Average value of the 60-day rolling beta for ARKF is {arkf_60_beta_ave}")
```

```python
arkf_60_beta.plot(figsize=(15,7), title="60-day rolling beta for ARKF")
```
